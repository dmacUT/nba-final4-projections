```python
class BinaryTreeNode:

    def __init__(self, value):
        self.value = value
        self.left  = None
        self.right = None

    def insert_left(self, value):
        self.left = BinaryTreeNode(value)
        return self.left

    def insert_right(self, value):
        self.right = BinaryTreeNode(value)
        return self.right
```

```python
def is_super_balanced(root):
    depth_r = 0
    depth_l = 0
    nodes_r = 0
    nodes_l = 0
    
  
    
    elif root.left > 0:
        depth_l += 1
        nodes_l += 1
        is_super_balanced(root.left)
        
    if root.right == None:
    elif root.right > 0:
        depth_l += 1
        nodes_r += 1
        is_super_balanced(root.right)
    
    if abs(depth_r - depth_l) > 2:
        return False
    elif abs(nodes_r - nodes_l) <= 1:
        return True
    else:
        return False
        
        
```
