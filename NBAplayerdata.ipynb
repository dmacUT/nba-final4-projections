```python
years = ['2006','2007','2008','2009','2010','2011','2012','2013','2014','2015','2016','2017']
```

```python
for year in years:
    req = requests.get("https://www.basketball-reference.com/leagues/NBA_{}_advanced.html".format(year))
    content = req.content
    soup = BeautifulSoup(content)
    tables = soup.find('table')
    globals()['advanced_stats_%s' % year] = pd.read_html(str(tables))
    globals()['advanced_stats_%s' % year] = globals()['advanced_stats_%s' % year][0]
```
