```python
def sum_past_yrs(df):
    new = df
    for k, v in new.items():
        count = 1
        prior = k-count
        # 17 - prior (16)
        while prior > 4:
            # p, v in 15
            for player, votes in new[k].items():
                # p, in 14
                if player in new[prior].keys():
                    # p15 + p14
                    new[k][player] += new[prior][player]
            prior -= 1
    return new
```
